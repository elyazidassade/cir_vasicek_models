```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Modèle de taux d'intérêt à court terme

Les modèles de taux d'intérêt à court terme sont utilisé en mathématiques financières pour modéliser l'évolution des taux d'intérêt court terme. Ils ont pour but d'estimer les scénarios potentiels pour les fluctuations des taux d'intérêt dans le futur. Il existe deux types de modèles : les modèles one-factor (ex: modèle de Vasicek) et les modèles multi-factors (ex: modèle de CIR).


### Modèle de Vasicek

Il s'agit de la solution de l'équation différentielle stochastique (EDS) :

$$ dr_t = a (b - r_t)dt + \sigma dW_t $$ 

où : 

- b : moyenne de taux d'intérêt long terme
- a : vitesse de réversion
- $\sigma$ : volatilité instantanée du taux d'intérêt
- $(W_t)_{t \geq 0}$ est un processus de Wiener suivant la loi normale $N(0,dt)$



### Modèle Cox-Ingresoll-Ross (CIR)

Il s'agit de la solution de l'équation différentielle stochastique (EDS) :

$$ dr_t = \kappa (\mu - r_t)dt + \sigma \sqrt r_t dW_t $$ 

où : 

- $\mu$ : moyenne de taux d'intérêt long terme
- $\kappa$ : vitesse de réversion
- $\sigma$ : volatilité instantanée du taux d'intérêt
- $(W_t)_{t \geq 0}$ est un processus de Wiener suivant la loi normale $N(0,dt)$

Cette méthode, contrairement au modèle de Vasicek, implémente une diffusion en racine carrée du taux d'intêret et ne génère pas de taux négatifs.

NB : le taux r0 introduit en paramètre est annuel. Il faut le convertir en taux instantané. De même, les taux calculés sont instantanés et doivent être annualisés.

```python
def inst_to_ann(r):
    """
    Convertir un taux d'intérêt instantané à un taux d'intérêt annuel
    """
    return np.exp(1-r)

def ann_to_inst(r):
    """
    Convertir un taux d'intérêt annuel à un taux d'intérêt instantané
    """
    return np.log(1+r)
```

### Simulation du taux d'intêret par le modèle Vasicek pour un certain nombre de scénarios entré en paramètre

```python
def vasicek(a, b, T, sigma, N, n_scenarios):
    
    dt = T/N #time increment
    aleatoire = np.random.normal(0, scale=np.sqrt(dt), size=(N, n_scenarios))
    #each column is a scenario, each row is a period
    taux = np.empty_like(aleatoire)
    
    a = ann_to_inst(a)
    taux[0] = a
    
    for i in range(1,N):
        r_t = (taux[i-1])
        dr_t = a*(b-r_t)*dt + sigma*aleatoire[i]
        taux[i] = r_t +dr_t
        
    return pd.DataFrame(inst_to_ann(taux), index=np.arange(0,T-dt,dt))
```

```python
vasicek_simulation = vasicek(a=0.03, b=0.05, T=1, sigma=0.05, N=253, n_scenarios=250)
```

```python
vasicek_simulation.plot(figsize=(15,7), legend=False, title='short term rate interest - 250 simulated scenarios by Vasicek model')
```

```python
vasicek_simulation.head()
```

```python
mean_rates_per_period_vasicek = vasicek_simulation.mean(axis='columns') #r.mean(axis=1)
mean_rates_per_period_vasicek.plot(figsize=(15,8))
```

### Simulation du taux d'intêret par le CIR pour un certain nombre de scénarios entré en paramètre

```python
def cir(r0=0.02, T=1, kappa=2, mu=0.03, sigma=0.01, N=253, n_scenarios=1):

    dt = T/N #time increment
    aleatoire = np.random.normal(0, scale=np.sqrt(dt), size=(N, n_scenarios))
    #each column is a scenario, each row is a period
    taux = np.empty_like(aleatoire)
    
    r0 = ann_to_inst(r0)
    taux[0] = r0
    
    for i in range(1,N):
        r_t = taux[i-1]
        dr_t = kappa*(mu-r_t)*dt + sigma*np.sqrt(r_t)*aleatoire[i]
        taux[i] = r_t +dr_t
        
    return pd.DataFrame(inst_to_ann(taux), index=np.arange(0,T-dt,dt))
```

```python
cir_simulation = cir(r0=0.03, T=1, kappa=0.1, mu=0.03, sigma=0.1, N=253, n_scenarios=250)
```

```python
cir_simulation.head()
```

```python
cir_simulation.plot(figsize=(15,7), legend=False, title='short term rate interest - 250 simulated scenarios by CIR model')
```

### Moyenne pour chaque période

```python
mean_rates_per_period = cir_simulation.mean(axis='columns') #r.mean(axis=1)
mean_rates_per_period.plot(figsize=(15,8), title='mean of the 250 simulated scenarios by CIR model')
```
